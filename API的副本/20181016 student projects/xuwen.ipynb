```python
events=[]
for e in data["features"]:
    mag = e['properties']['mag']
    time = e['properties']['time']
    lon = e['geometry']['coordinates'][0]
    lat = e['geometry']['coordinates'][1]
    events.append([time, lon, lat, mag])
```
