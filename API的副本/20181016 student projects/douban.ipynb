```python
events=[]
for e in data['subjects']:
    title=e['title']
    rating=e['rating']['average']
    events.append([title,rating])
```

```python
events
```

```python
import csv
f=open('douban.csv','w',newline='')
writer=csv.writer(f)
header=['title','rating']
writer.writerow(header)
writer.writerows(events)
f.close()
```
